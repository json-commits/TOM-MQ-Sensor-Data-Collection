```python
import pandas as pd
red_data = pd.DataFrame()
for i in range(1, 9):
    print(f"reading Red{i}.xlsx...", end=" ")
    red_data = pd.concat([red_data, pd.read_excel(f'Data (12-09-2023)/Red{i}.xlsx', index_col=0)])
    print("done")
red_data['data_label'] = 'red'
red_data
```

```python
breaker_data = pd.DataFrame()
for i in range(4, 14):
    if i == 10:
        continue
    print(f"reading Breaker{i}.xlsx...", end=" ")
    breaker_data = pd.concat([breaker_data, pd.read_excel(f'Data (12-09-2023)/Breaker{i}.xlsx', index_col=0)])
    print("done")
breaker_data['data_label'] = 'breaker'
breaker_data
```

```python
turning_data = pd.DataFrame()
for i in range(3, 11):
    print(f"reading Turning{i}.xlsx...", end=" ")
    turning_data = pd.concat([turning_data, pd.read_excel(f'Data (12-09-2023)/Turning{i}.xlsx', index_col=0)])
    print("done")
turning_data['data_label'] = 'turning'
turning_data
```

```python
import pandas as pd
clean_data = pd.DataFrame()

for i in ['Clean data 12-6-23.xlsx', 'clean2.xlsx']:
    print(f"reading {i}...", end=" ")
    clean_data = pd.concat([clean_data, pd.read_excel(f'Data (12-09-2023)/{i}', index_col=0)])
    print("done")
clean_data['data_label'] = 'clean'
clean_data
```

```python
import seaborn as sns
master_df = pd.concat([red_data, breaker_data, turning_data, clean_data])
master_df
```

```python
# master_df = pd.concat([red_data, breaker_data, turning_data])

A0_plot = sns.lmplot(data=master_df, x='time_A', y='A1', row='data_label', legend='full', markers='x')
# for df in red_dfs[1:]:
#     A0_plot = sns.scatterplot(data=df, x='time_A', y='A0', ax=A0_plot)
    
A0_plot
```
